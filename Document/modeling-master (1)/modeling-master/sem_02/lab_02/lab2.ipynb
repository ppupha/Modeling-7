# Равномерное распределение

## Плотность распределения

$
f_X (x) =
\begin{cases}
  \frac{1}{b-a}, x \in [a,b] \\
  0, x \notin [a, b] \\
\end{cases}
$

## Функция распределения

$
F_X (x) =
\begin{cases}
  0, x < a \\
  \frac{x - a}{b - a}, a \le x < b \\
  1, x \geq b \\
\end{cases}
$

```
def udDensity(a: float, b: float, x: float) -> float:
    return 1 / (b - a) if a <= x <= b else 0
```

```
def udFunction(a : float, b : float, x : float) -> float:
    return (x - a) / (b - a) if a <= x < b else 0 if x < a else 1
```

```
a = float(input("a: "))
b = float(input("b: "))
diff = b - a
x = np.arange(a - diff / 2, b + diff / 2, 0.001)
yDensity  = [udDensity(a, b, _x) for _x in x]
yFunction = [udFunction(a, b, _x) for _x in x]
```

```
plt.plot(x, yDensity)
plt.title('Плотность равномерного распределения', fontsize=15)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```

```
plt.plot(x, yFunction)
plt.title('Функция равномерного распределения', fontsize=15)
plt.xlabel('x')
plt.ylabel('F(x)')
plt.grid(True)
plt.show()
```

# Гауссово распределение

## Плотность распределения

$
f_X (x) = \frac{1}{\sigma \sqrt{2 \pi}} e^{-\frac{(x - \mu)^2}{2 \sigma^2}}
$

## Функция распределения

$
\Phi (x) = \frac{1}{\sigma \sqrt{2\pi}} \int_{-\infty}^x e^{-\frac{(t- \mu)^2}{2 \sigma^2}} dt
$

```
from scipy.stats import norm
```

```
sigma = float(input("sigma: "))
mu = float(input("mu: "))
x = np.arange(-(sigma + mu) * 2, (mu + sigma) * 2, 0.001)
yDensity = norm.pdf(x)
yFunction = norm.cdf(x)
#yDensity  = [gDensity(sigma, mu, _x) for _x in x]
#yFunction  = [gFunction(sigma, mu, _x) for _x in x]
```

```
plt.plot(x, yDensity)
plt.title('Плотность распределения Гаусса', fontsize=15)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```

```
plt.plot(x, yFunction)
plt.title('Функция распределения Гаусса', fontsize=15)
plt.xlabel('x')
plt.ylabel('F(x)')
plt.grid(True)
plt.show()
```
